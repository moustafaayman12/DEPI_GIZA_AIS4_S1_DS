# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  
    
```

```python
number = int(input("Enter a number to calculate its factorial: "))

factorial = 1
steps = []

while number > 1:
    factorial *= number
    steps.append(factorial)
    number -= 1

print("Steps:", steps)
print("Factorial:", factorial)
```

# Create a List of Squares of Numbers from 1 to 10
## Example
Output: [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]

```python
# Write your answer here
list_sqr=[]


def squarse (n):
    sqr=n*n
    list_sqr.append(sqr)
    return squarse(n-1) if n>=1 else list_sqr
squarse(10).reverse()
print(list_sqr)
```

**Another Solution**

```python
# Write your answer here 
  
def squarse_1(n):
    list_sqr = []
    while n >= 1:
        sqr = n * n
        list_sqr.append(sqr)
        n -= 1
    return list_sqr
result = squarse_1(10)
result.reverse() 
print(result)   
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here

def even_numbers(n):
    sum_even = 0
    for even in n:
        if even % 2 == 0:
            sum_even += even
    return sum_even
print(even_numbers([1, 2, 3, 4, 5, 6]))
           
           
    
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
# Write your answer here
numbers =[5, 2, 7, 3, 1]
numbers.sort()
def median(n):
    length = len(n)
    if length % 2 == 0:
        mid1 = n[length // 2 - 1]
        mid2 = n[length // 2]
        median_value = (mid1 + mid2) / 2
    else:
        median_value = n[length // 2]
    return median_value
print("Median:", median(numbers))
```
