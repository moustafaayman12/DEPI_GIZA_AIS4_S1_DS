```python

frist_number= float(input("Enter first number: "))
opretor=input("Enter operator : ")
second_number= float(input("Enter second number: "))

if opretor=='+':
    print(frist_number + second_number)
elif opretor=='-':
    print(frist_number - second_number)
elif opretor=='*':
    print(frist_number * second_number)
elif opretor=='/':
    print(frist_number / second_number)
elif opretor=='%':
    print(frist_number % second_number)
else:
    print("Invalid operator")                
```
