```python
#task one
class process_in_email:
   
        
    def __init__(self, email):
        self.email = email
    def get_username(self):
        username = self.email.split("@")[0]
        print("Username:", username)
    def get_domain(self):
        domain= self.email.split("@")[1]
        domain= domain.split(".")[0]
        print("Domain:", domain)
        
    def validate_email(self,email):  
        email_split = self.email.split("@")
       
       
        if email.count("@") == 1 :
            
            if email.count(".") >= 1 :
                if "."in email_split[1]:
                    print("Valid email address.")
                else:
                    print("Invalid email address.")
                    
            else:
                print("Invalid email address.")        
        else:
            print("Invalid email address.")            
                    
        if email.endswith(".com"):
            print("commercial domain")    
        elif email.endswith(".edu"):
            print("edution domain")    
        else:
            print("other domain")
            
            
                   
email_1 = process_in_email("mostafa@gmail.edu")
email_1.validate_email(email_1.email)
email_1.get_username()
email_1.get_domain()
```

```python
#task three
message = "&&&**$gnirtS PLIO!!@1234"
massage_slice =message[6:17]
print(massage_slice)
message_slice_split1=massage_slice.split(" ")
print(message_slice_split1[0])
reversed_message2=message_slice_split1[0][::-1]
print(reversed_message2) 
first_word=reversed_message2
scound_word="PLIO"
def shifted_vowels(x):
    word = x
    
    for char in word:
        if char == "I":
            word = word.replace("I", "E")
        elif char == "O":
            word = word.replace("O", "U")
        
    return word

scound_word=shifted_vowels(scound_word)

decode=" ".join(first_word + " " + scound_word)
print(f"decoded message: {decode}")
       
       
    
 
           
               

```
