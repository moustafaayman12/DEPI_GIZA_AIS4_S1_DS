```python
class calculator:
    def __init__(self, a: int, b: int):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b 
    def subtract(self):
        return self.a - self.b
    
x1= calculator(10, 5)
print(x1.add())
print(x1.subtract())      
```

```python
from abc import ABC, abstractmethod

class animal(ABC):
    @abstractmethod
    def sound(self):
        return "wwwww"
class dog(animal):
    def sound(self):
        return "bark"

    def eat(self):
        
        return "eating"

class cat(animal):
    def sound(self):
        return "meow"
   
x=dog()
print(x.sound()) 
print(x.eat())
y=cat()
print(y.sound())
```
