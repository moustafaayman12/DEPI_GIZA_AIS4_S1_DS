```python
# assignment
import os
import random


def create_file_folder(path):
    length=0
    if not os.path.exists(path):
        os.mkdir(path)
        for i in range(0, 20):
            new_file =os.path.join(path, f"depi_{i}.txt")
            with open(new_file, 'a') as f:
                pass
            length+=1
        print(f"length of created files inside the folder: {length}")    
    else:
        print("folder already exists")        
    
create_file_folder("C:/dpei1/DEPI_GIZA_AIS4_S1_DS/python/Section_oop_4/depi_folder") 

def delete_half_files_randomly(path):
    if os.path.exists(path):
        files = os.listdir(path)
        length=len(files)
        num_files_to_delete = len(files) // 2
        files_to_delete = random.sample(files, num_files_to_delete)
        
        for file_name in files_to_delete:
            file_path = os.path.join(path, file_name)
            os.remove(file_path)
            length-=1
        
        print(f"Deleted {num_files_to_delete} files randomly from the folder.")
        print(f"Remaining files in the folder: {length}")
    else:
        print("folder does not exist")
delete_half_files_randomly("C:/dpei1/DEPI_GIZA_AIS4_S1_DS/python/Section_oop_4/depi_folder")    


   
```
